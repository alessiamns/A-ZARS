```python
import pandas as pd
import numpy as np
import sys
import codecs
import nltk

df_reviews = pd.read_json ("data/reviews_with_language.json")
df_reviews = df_reviews.reset_index()
```

# GroupByReviews per LINGUA

```python
all_languages = df_reviews.Language.unique()
grouped = df_reviews.groupby('Language')

for group in grouped:
    #print(group)
    for language in all_languages:
        a = grouped.get_group(language)
        apds = pd.DataFrame(a)
        #print(apds['Language'].value_counts())
        #apds.to_json('data/rev/rev_'+language+'.json')
```

## GroupByReviews per NUMERO STRUTTURE

```python
all_languages = df_reviews.Language.unique()
grouped = df_reviews.groupby('Language')

for group in grouped:
    #print(group)
    for language in all_languages:
        a = grouped.get_group(language)
        apds = pd.DataFrame(a)
        print(language, len(apds['Name'].unique()))
        #apds.to_json('data/rev/rev_'+language+'.json')
```

## GroupByReviews per PROVENIENZA

```python
all_nations = df_reviews.Hometown.unique()
grouped = df_reviews.groupby('Hometown')

for group in grouped:
    #print(group)
    for nat in all_nations:
        a = grouped.get_group(nat)
        apds = pd.DataFrame(a)
        print(nat, apds['Hometown'].value_counts())
        #apds.to_json('data/rev/rev_'+language+'.json')
    
```
