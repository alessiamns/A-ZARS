```python
import pandas as pd
import numpy as np
```

```python
df_reviews = pd.read_json ("data/reviews.json",lines=True)
df_geo = pd.read_json ("data/country_geonames_id.json")
df_geo['country'].unique()
```

```python
n = df_reviews.shape[0]
geonames_id = np.zeros(n)
df_reviews['geonames_id'] = geonames_id
```

```python
for i in range(0,len(df_geo)):
    print(df_geo['country'][i])
    print(df_geo['geonames_id'][i])
    m = df_reviews['Hometown'] == df_geo['country'][i]
    df_reviews.loc[m, 'geonames_id'] = df_geo['geonames_id'][i]
df_reviews.to_json('data/reviews_with_geo1.json')
```
